### Regression analysis on the business license data and stock market (high and low)
##### High price: there is significant relationship
* P-value=0, t-value=3.568, because of this small p-value and large t-value we conclude there is significnat relationship between business license application and high stock market price.
* When the stock market price goes up, we can see the business license application goes up by 0.0007

##### Low price: there is significant relationship 
* * P-value=0, t-value=3.666, because of this small p-value and large t-value we conclude there is significnat relationship between business license application and low stock market price.
* When the stock market price goes up, we can see the business license application also goes up by 0.0007


```python
ols = smf.ols(formula = "COUNT ~ High",
              data = test4)
model = ols.fit()
model.summary()
```

```python
ols = smf.ols(formula = "COUNT ~ Low",
              data = test4)
model = ols.fit()
model.summary()
```

#### Same as we see from the regression above, from this graph, we also observe there is significant relationship between business license application and high stock market price, as the regression line points to down. 
* We can tell there is positive relationship between business license application and high stock market price
* As the stock market price goes up, there is more business license application. Becuase the coefficient is small, the relationship is not obvious


```python
m = pd.Series(name = "COUNT", data = model.resid + test4["High"] * model.params["High"])
sns.regplot(x = test4["High"], y = m, lowess = True, 
            scatter_kws = {"alpha" : 0.1, "color" : "red"}, line_kws = {"color" : "grey"},);
```

#### Same as we see from the regression above, from this graph, we also observe there is significant relationship between business license application and low stock market price, as the regression line points to up. 
* We can tell there is positive relationship between business license application and low stock market price
* As the stock market price goes up, there is more business license application, however as the coefficient is small, the trend is not obvious

```python
n = pd.Series(name = "COUNT", data = model.resid + test4["Low"] * model.params["Low"])
sns.regplot(x = test4["Low"], y = n, lowess = True, 
            scatter_kws = {"alpha" : 0.1, "color" : "red"}, line_kws = {"color" : "grey"});
```

### We use Seaborn to double check the result, and we find same regression relationship as we found before
* Therefore, we conclude that there is negative relationship between high stock market price and number of license application, also there is positive relationship between low stock market price and number of license application

```python
f = pd.Series(name = "Component + Residual", data = model.resid + test4["High"] * model.params["High"])
ax = sns.regplot(x = test4["High"], y = f, scatter = True, fit_reg = False,
            scatter_kws = {"alpha" : 0.1, "color" : "red"})
sns.regplot(x = test4["High"], y = f, scatter = False, 
            lowess = False, robust = True, label = "Linear",
            scatter_kws = {"alpha" : 0.1, "color" : "red"}, line_kws = {"color" : "yellow", "linewidth" : 2})
sns.regplot(x = test4["High"], y = f, scatter = False, 
            lowess = True, label = "Lowess",
            scatter_kws = {"alpha" : 0}, line_kws = {"color" : "grey", "linewidth" : 2}, ax = ax);

ax.legend()
```

```python
f = pd.Series(name = "Component + Residual", data = model.resid + test4["Low"] * model.params["Low"])
ax2 = sns.regplot(x = test4["Low"], y = f, scatter = True, fit_reg = False,
            scatter_kws = {"alpha" : 0.1, "color" : "red"})
sns.regplot(x = test4["Low"], y = f, scatter = False, 
            lowess = False, robust = True, label = "Linear",
            scatter_kws = {"alpha" : 0.1, "color" : "red"}, line_kws = {"color" : "yellow", "linewidth" : 2})
sns.regplot(x = test4["Low"], y = f, scatter = False, 
            lowess = True, label = "Lowess",
            scatter_kws = {"alpha" : 0}, line_kws = {"color" : "grey", "linewidth" : 2}, ax = ax);

ax2.legend()
```

#### We run regression on the unemployment rate and stock market price, and found there is significant relationship
##### High stock market price: negative relationship 
* while holding other factors fixed, we see the p-value for the high price is 0.00, also we see the t-value is -10.809. So we conclude there is significant relationship between unemployment rate and high stock market price, because of the small P-value and large t-value
* when the high stock market price goes up, the unemployment rate goes down by 0.0008

##### Low stock market price: negative relationship 
* while holding other factors fixed, we see the p-value for the low price is 0.00, also we see the t-value is -10.025. So we conclude there is significant relationship between unemployment rate and low stock market price, because of the small P-value and large t-value
* when the low stock market price goes down, the unemployment rate goes up by 0.0008


```python
ols = smf.ols(formula = "Chicago ~ High",
              data = test5)
model = ols.fit()
model.summary()
```

```python
ols = smf.ols(formula = "Chicago ~ Low",
              data = test5)
model = ols.fit()
model.summary()
```

#### We run regression on unemployment rate and number of license application: No significance relationship


```python
ols = smf.ols(formula = "Chicago ~ COUNT",
              data = test5)
model = ols.fit()
model.summary()
```

#### We hold fix the count, the relationship is still strong for high price and unemployment rate

```python
ols = smf.ols(formula = "Chicago ~ High + COUNT",
              data = test5)
model = ols.fit()
model.summary()
```

#### We plot the unemployment and number of license application and comfirmed there is no significant relationship between these two data

```python
m = pd.Series(name = "Chicago", data = model.resid + test5["COUNT"] * model.params["COUNT"])
sns.regplot(x = test5["COUNT"], y = m, lowess = True, 
            scatter_kws = {"alpha" : 0.1, "color" : "red"}, line_kws = {"color" : "grey"});
```
